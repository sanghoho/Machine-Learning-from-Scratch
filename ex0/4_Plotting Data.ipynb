```octave
t = [0:0.01:0.98]
```

```octave
y1 = sin(2*pi*4*t);
plot(t, y1)
```

```octave
y2 = cos(2*pi*4*t);
plot(t, y2)
```

# Show Multiple Graph


## `hold on`
- `hold on`을 쓰세요

```octave
plot(t, y1)
hold on;
plot(t, y2, 'rx')

xlabel("time")
ylabel("value")

legend('sin', 'cos')
title('my plot')

print -dpng 'myPlot.png' % Plot 저장
```

명령 prompt에서 `close`를 치면 plot 창이 닫힌다.

## `figure`

```octave
figure(1); plot(t, y1)

figure(2); plot(t, y2)
```

## subplot

```octave
subplot(1, 2, 1); % Divides plot a 1x2 grid, access first element
plot(t, y1)
subplot(1, 2, 2)
plot(t, y2)

axis([0.5 1 -1 1]) % x축 ~ y축 설정 

% clf; % subplot의 plot을 지워줌
```
